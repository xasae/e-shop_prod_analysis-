## Q1

### Na jaké kategorii produktů máme největší obrat? A zajímalo by mě i jestli se to v jednotlivých měsících mění.

Přidá dva sloupce do 'all_data': 'Revenue' a 'Month'.
Adds two columns to 'all_data': 'Revenue' and 'Month'

```python
all_data['Date'] = pd.to_datetime(all_data['Date'])
all_data['Month'] = all_data['Date'].dt.month
all_data['Revenue'] = (all_data['Quantity'] * all_data['Price']).astype(float)
all_data
```

Definuje, která kategorie má největší obrat

Defines which category has the highest revenue

```python
each_cat_total = all_data.groupby('Category')['Revenue'].sum().reset_index()
each_cat_total.rename(columns={'Revenue': 'Total'}).sort_values(by=['Total'], ascending=False)
```

Zobrazuje změny pro každý měsíc

Shows changes for each month

```python
max_cat_revenue = all_data.groupby(['Month', 'Category'])['Revenue'].sum().reset_index()
max_cat_revenue_pivot = max_cat_revenue.loc[max_cat_revenue.groupby('Month')['Revenue'].idxmax()]\
    .pivot(
        columns='Month',
        index='Category',
        values='Revenue'
    )
max_cat_revenue_pivot
```

Kategorie "Televize" má největší obrat

Category 'Televize' has the highest revenue

## Q2

### Který den v týdnu je nejsilnější na počet objednávek?

Přidá nový sloupec 'DayOfWeek' a definuje den v týdnu.

Adds new column 'DayOfWeek' and defines the day of the week

```python
all_data['DayOfWeek'] = all_data['Date'].dt.day_name()
all_data
```

Počítá počet produktů v transakci uique a třídí je podle dnů v týdnu.

Counts number of products in the uique transaction and sorts by day of the week

```python
products_in_transaction = all_data.groupby('Transaction ID')['DayOfWeek'].value_counts().reset_index()
products_in_transaction.rename(columns={'count': 'Products'}, inplace=True)
products_in_transaction
```

Počítá počet transakcí za den

Counts number of transactions by day

```python
max_transactions_by_days = products_in_transaction.groupby('DayOfWeek')['Transaction ID'].count().reset_index()
max_transactions_by_days.rename(columns={'Transaction ID': 'Transactions'}, inplace=True)
max_transactions_by_days.sort_values(by=['Transactions'], ascending=False)
```

Počítá počet produktů za den

Counts number of products by day

```python
max_products_by_day = all_data['DayOfWeek'].value_counts().reset_index()
max_products_by_day.rename(columns={'count': 'products'}, inplace=True)
max_products_by_day
```

Většina transakcí a produktů probíhá v sobotu

Most transactions and products are on Saturday

## Q3

### Která kategorie se prodává nejčastěji spolu s produkty z kategorie Televize (resp. jsou spolu v jedné objednávce)? 

```python
transaction_cat = all_data.get(['Transaction ID', 'Category'])
transaction_cat
```

Získání ID transakce s kategorií 'Televize' 

Getting transaction id with the category 'Televize' 

```python
transactions_tele = transaction_cat[transaction_cat['Category'].isin(['Televize'])]
ids = transactions_tele['Transaction ID']
print(list(ids))
```

```python
transactions_ids = transaction_cat[transaction_cat['Transaction ID'].isin(ids) & (transaction_cat['Category']!='Televize')]
transactions_ids['Category'].value_counts().reset_index()
```

Kategorie "Audio" se nejčastěji prodává společně s produkty z kategorie "Televize".

Category 'Audio' is most often sold together with products from the category 'Televize'

## Q4

### Od 18. 3. 2022 jsem klukům z marketingu navýšil budget na online marketingových platformách (Google Ads, Sklik, Facebook). Dokážeš mi říct, jestli to vedlo k nějaké změně v prodeji?

```python
sales_per_day = all_data.groupby('Date')['Revenue'].sum().reset_index()
sales_per_day
```

```python
after_march = sales_per_day.loc[sales_per_day['Date'] >= '2022-03-18']
befor_march = sales_per_day.loc[sales_per_day['Date'] < '2022-03-18']
after_march['Revenue'].sum() - befor_march['Revenue'].sum()
```

Příjmy se zvýšily na 318250 po 18. březnu

Revenue increased on 318250 after March 18th
